```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get('https://klimi.mk/')
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
proizvoditeli_elementi = soup.select('.proizvoditel')
proizvoditeli_elementi
```

```python
proizvoditeli = [
    proizvoditeli_elementi.get('id') for proizvoditeli_elementi in proizvoditeli_elementi
]
proizvoditeli
```

```python
slider_element = soup.select_one('#slider')
print(slider_element)
```

```python
sliki_elements = slider_element.select('img.img.img-fluid.slika-klima-slajder')
print(len(sliki_elements))
print(sliki_elements)
```

```python
slika_urls = [sliki_elements.get('src') for sliki_elements in sliki_elements]
print(len(slika_urls))
print(slika_urls)
```
